```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
        
        result1 = switch1 & switch2
        result2 = switch1 or switch2
        result3 = not(switch1 or switch2)
        result4 = not(switch1 & switch2)
        
        print(result4)
        print(result2)
        if result1 == 1:
            base.leds[0].on()
            print ('output = and true')
            
        else:
            base.leds[0].off()
            
        if result2 == 1:
            base.leds[1].on()
            print("output= or true")
            
        else:
            base.leds[1].off()
            
        if result3 == 1:
            base.leds[2].on()
            print("output= exor true")
            
        else:
            base.leds[2].off()

        if result4 == 1:
            base.leds[3].on()
            print("output= not true")
        
        else:
            base.leds[3].off()
            print("output=false (invalid)")

print("output=false (invalid)")
main()
```
