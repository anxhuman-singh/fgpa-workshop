# Half Adder

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch1 = base.switches[1].read()
        switch2 = base.switches[0].read()

        sum = switch1 ^ switch2
        carry = switch1 and switch2
        if sum == 1:
            base.leds[0].on()
            print('sum = 1')
        else:
            base.leds[0].off()

        if carry == 1:
            base.leds[1].on()
            print('carry = 1')
        else:
            base.leds[1].off() 
```

# Half Subtractor

```python
from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    #and logic code 
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
 
        difference =  switch1^switch2
        borrow = (~(switch1)&1)&switch2
 
        if difference==1:
            base.leds[0].on()
            print ('sum_output = 1')
 
        else:
            base.leds[0].off()
            print("Sum output = 0")
 
        if borrow== 1:
            base.leds[1].on()
            print("carry= 1")
 
        else:
            base.leds[1].off()
            print("carry = 0")
main()
```

from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    counter = 0
    for i in range(0,16):
        bit1 = counter&1
        bit2 = (counter>>1)&1
        bit3 = (counter>>2)&1
        bit4 = (counter>>3)&1
 
        if bit1==1:
            base.leds[0].on()
        else:
            base.leds[0].off()
 
 
        if bit2==1:
            base.leds[1].on()
        else:
            base.leds[1].off()
 
 
        if bit3==1:
            base.leds[2].on()
        else:
            base.leds[2].off()
 
 
        if bit4==1:
            base.leds[3].on()

            base.leds[3].off()
 
        print(counter)

main()
