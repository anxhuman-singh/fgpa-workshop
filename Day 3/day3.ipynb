# Half Adder

```python
from time import sleep
from pynq.overlays.base import BaseOverlay

def main():
    base = BaseOverlay('base.bit')
    # LED access
    
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
    
    #and logic code 
    while True:
        switch1 = base.switches[1].read()
        switch2 = base.switches[0].read()

        sum = switch1 ^ switch2
        carry = switch1 and switch2
        if sum == 1:
            base.leds[0].on()
            print('sum = 1')
        else:
            base.leds[0].off()

        if carry == 1:
            base.leds[1].on()
            print('carry = 1')
        else:
            base.leds[1].off() 
```

# Half Subtractor

```python
from time import sleep
from pynq.overlays.base import BaseOverlay
 
def main():
    base = BaseOverlay('base.bit')
    # LED access
 
    for led in base.leds:
        led.on()
    sleep(0.25)
    for led in base.leds:
        led.off()
    sleep(1)
 
    #and logic code 
    while True:
        switch1 = base.switches[0].read()
        switch2 = base.switches[1].read()
 
        difference =  switch1^switch2
        borrow = (~(switch1)&1)&switch2
 
        if difference==1:
            base.leds[0].on()
            print ('sum_output = 1')
 
        else:
            base.leds[0].off()
            print("Sum output = 0")
 
        if borrow== 1:
            base.leds[1].on()
            print("carry= 1")
 
        else:
            base.leds[1].off()
            print("carry = 0")
main()
```
