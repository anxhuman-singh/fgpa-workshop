# Shift Register


```python
from pynq.overlays.base import BaseOverlay
from time import sleep

base = BaseOverlay('base.bit')

state = [0, 0, 0, 0]

def show_state():
    for i in range(4):
        base.leds[i].write(state[i])
    print('State:', *state)

def wait_release(button_index):
    while base.buttons[button_index].read():
        sleep(0.05)

show_state()
print('Shift register running...')

while True:
    if base.buttons[0].read():
        state = [0, state[0], state[1], state[2]]
        show_state()
        wait_release(0)

    elif base.buttons[1].read():
        state = [state[1], state[2], state[3], 1]
        show_state()
        wait_release(1)

    elif base.buttons[2].read():
        state = [state[1], state[2], state[3], 0]
        show_state()
        wait_release(2)

    elif base.buttons[3].read():
        state = [state[1], state[2], state[3], 0]
        show_state()
        wait_release(3)

    sleep(0.05)
```
